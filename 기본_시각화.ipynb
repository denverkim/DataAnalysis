# 선그래프
주어진 월별 온도 데이터를 이용해 선그래프를 그리세요.

image.png

```python
import matplotlib.pyplot as plt

months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
temperature = [-5, -3, 2, 10, 15, 20, 25, 24, 18, 10, 3, -2]

plt.plot(months, temperature, marker='o')
plt.title('월별 평균 기온')
plt.xlabel('월')
plt.ylabel('기온 (°C)')
plt.grid(True)
plt.show()
```

# 막대그래프

```python
import matplotlib.pyplot as plt

# 데이터 설정
months = ['1월', '2월', '3월', '4월', '5월', '6월']
sales = [120, 150, 180, 220, 250, 300]

# 막대그래프 그리기
plt.bar(months, sales, color='skyblue')  # 막대 색상은 'skyblue'로 설정

# 그래프 제목 및 축 레이블
plt.title('월별 판매량')
plt.xlabel('월')
plt.ylabel('판매량 (개수)')

# 그래프 보여주기
plt.show()
```
