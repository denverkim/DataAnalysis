```python
import pandas as pd
```

```python
left = pd.DataFrame({ 'id':[1,2,3,4,5],
                     'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                     'subject_id':['sub1','sub2','sub4','sub6','sub5']})
left
```

```python
right = pd.DataFrame( {'id':[1,2,3,4,5],
                       'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
                       'subject_id':['sub2','sub4','sub3','sub6','sub5']})
right
```

```python
pd.merge(left, right, on='subject_id', how='left')
```

```python
pd.merge(left, right, on='subject_id', how='right')
```

```python
pd.merge(left, right, on='subject_id', how='outer')
```

```python
pd.merge(left, right, on='subject_id', how='inner')
```

```python
import pandas as pd
left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})
left
```

```python
right = pd.DataFrame(
         {'id':[1,2,3,4,5],
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})
right
```

```python
pd.merge(left,right,on=['id','subject_id'])
```

# Concat

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df1
```

```python
df2 = pd.DataFrame([['c', 3], ['d', 4]], columns=['letter', 'number'])
df2
```

```python
pd.concat([df1, df2]) #위아래로 concat
```

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df1
```

```python
df2 = pd.DataFrame([['bird', 'polly'], ['monkey', 'george']], columns=['animal', 'name'])
df2
```

```python
pd.concat([df1, df2], axis=1)  #옆으로 concat
```
