```python
# 다음 파일을 읽고 데이터 병합을 수행 하시오 (https://github.com/sehakflower/data/blob/main/titanic_1309.xlsx?raw=true).

import pandas as pd

titanic = pd.read_excel('https://github.com/sehakflower/data/blob/main/titanic_1309.xlsx?raw=true')
titanic.head()
```

# 데이터 붙이기

```python
# 상위 1000행을 train으로 나머지 행을 test로 저장
train = titanic.iloc[:1000]
test = titanic.iloc[1000:]
```

```python
# train, test 위아래로 붙이기
pd.concat([train, test]).head()
```

```python
# train, test 옆으로 붙이기
pd.concat([train, test], axis=1).head()
```

# 데이터 병합

```python
# name, pclass, sex, age를 df1(상위 1000행)으로 name, survived를 df2(하위 500행)로 저장
df1 = titanic[:1000][['name', 'pclass', 'sex', 'age']]
df2 = titanic[-500:][['name', 'survived']]
```

```python
# name 으로 inner join
print(pd.merge(df1, df2, on='name').shape)
pd.merge(df1, df2, on='name').head()
```

```python
# name 으로 outer join
print(pd.merge(df1, df2, on='name', how='outer').shape)
pd.merge(df1, df2, on='name', how='outer').head()
```

```python
# name 으로 left join
print(pd.merge(df1, df2, on='name', how='left').shape)
pd.merge(df1, df2, on='name', how='left').head()
```

```python
# name 으로 right join
print(pd.merge(df1, df2, on='name', how='right').shape)
pd.merge(df1, df2, on='name', how='right').head()
```
