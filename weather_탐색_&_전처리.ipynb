# 1-3-1 데이터 탐색 연습문제

```python
# Read weather.txt file into a dataframe.
import pandas as pd
df = pd.read_csv('weather.txt')
df
```

```python
# Print first 5 or last 3 rows of df 처음 다섯줄, 마지막 세줄 출력
print(df.head())
print(df.tail(3))
```

```python
# Get data types, index, columns, values 데이터타입, 인덱스, 컬럼들, 값들
print(df.dtypes)
print(df.index)
print(df.columns)
print(df.values)
```

```python
# Statistical summary of each column 각 컬럼에 대한 통계요약
df.describe()
```

```python
# Sort records by any column (descending order) 한 컬럼을 정해 내림차순으로 정렬
df.sort_values(by='avg_precipitation', ascending=False)
```
